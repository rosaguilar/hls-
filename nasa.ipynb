```python
def load_raster(path, crop=None):
    with rasterio.open(path) as src:
        img = src.read()

        # load first 6 bands
        img = img[:6]

        img = np.where(img == NO_DATA, NO_DATA_FLOAT, img)
        if crop:
            img = img[:, -crop[0]:, -crop[1]:]
    return img

def enhance_raster_for_visualization(raster, ref_img=None):
    if ref_img is None:
        ref_img = raster
    channels = []
    for channel in range(raster.shape[0]):
        valid_mask = np.ones_like(ref_img[channel], dtype=bool)
        valid_mask[ref_img[channel] == NO_DATA_FLOAT] = False
        mins, maxs = np.percentile(ref_img[channel][valid_mask], PERCENTILES)
        normalized_raster = (raster[channel] - mins) / (maxs - mins)
        normalized_raster[~valid_mask] = 0
        clipped = np.clip(normalized_raster, 0, 1)
        channels.append(clipped)
    clipped = np.stack(channels)
    channels_last = np.moveaxis(clipped, 0, -1)[..., :3]
    rgb = channels_last[..., ::-1]
    return rgb
```

```python
def getxy_topleft(df, regionid):
  # use year 2023 as reference
  row = df[df.id==f"{regionid}_2023"]
  coordinate_str = row.topleft.values[0].strip("()")
  # Split the string to separate x and y values
  x_str, y_str = coordinate_str.split(', ')
  # Convert the extracted strings to floats
  minx, maxy = float(x_str), float(y_str)
  return minx, maxy
```

```python
import rasterio
from rasterio.transform import from_origin


def save_raster(idregio, data, drive_folder, df):

  # Example data (replace with your actual numpy array)
  minx, maxy = getxy_topleft(df, idregio)

  # Define the transform (top-left corner x, y and pixel size)
  transform = from_origin(west=minx, north=maxy, xsize=30, ysize=30)

  # Define the CRS (use your appropriate CRS)
  crs = "EPSG:32618"  # UTM returned from splitter

  # Define the output path
  output_path = os.path.join(drive_folder, f"{idregio}_cropclass.tif")
  # Write the data to a GeoTIFF
  with rasterio.open(
      output_path,
      'w',
      driver='GTiff',
      height=data.shape[0],
      width=data.shape[1],
      count=1,  # Number of bands
      dtype=data.dtype,
      crs=crs,
      transform=transform,
  ) as dst:
      dst.write(data, 1)  # Write the data to the first band
```
